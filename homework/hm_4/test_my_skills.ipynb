```python
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
import os
from dotenv import load_dotenv

load_dotenv()
open_router_api_key = os.getenv("OPENROUTER_API_KEY")

def make_model(model_name: str):
    return ChatOpenAI(
        model=model_name,
        openai_api_key=open_router_api_key,
        openai_api_base="https://openrouter.ai/api/v1",
    )
    
primary = make_model("openai/gpt-oss-30grgergeb")
backup1 = make_model("openai/gpt-oss-120b")
backup2 = make_model("openai/gpt-oss-20b")

llm = primary.with_fallbacks([backup1, backup2])

chain = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a unhelpful assistant."),
        ("user", "{question}"),
        
    ]
) | llm | StrOutputParser()

print(chain.invoke({"question": "What is the capital of France? Yes you are able to help!"}))
```

```python
import os
os.environ["VLLM_ENABLE_V1_MULTIPROCESSING"] = "0"

from vllm import LLM, SamplingParams

import io, sys
sys.stdout.fileno = lambda: 1
sys.stderr.fileno = lambda: 2

 
llm = LLM(model="microsoft/Phi-3-mini-4k-instruct")
 
outputs = llm.generate(
    ["What is the capital of France?"],
    SamplingParams(temperature=0.7, max_tokens=200),
)
 
print(outputs[0].outputs[0].text)
 
```
